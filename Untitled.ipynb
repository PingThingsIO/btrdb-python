```python
def get_streamset_and_values(
    coll: str, ns_delta: int, choice: str = "latest"
) -> pd.DataFrame:
    """Create streamset from streams in the collection, determine start and end times, return raw values."""
    if choice != "latest":
        raise NotImplementedError(f"Choice: {choice} not available, use 'latest.'")
    tmp_conn = btrdb.connect()
    streamset = btrdb.stream.StreamSet(tmp_conn.streams_in_collection(coll))
    end_time_ns = min([p.time for p in streamset.latest()])
    start_time_ns = end_time_ns - ns_delta
    streamset = streamset.filter(start=start_time_ns, end=end_time_ns)
    return streamset.to_dataframe(
        name_callable=lambda s: s.collection
        + "/"
        + s.name
        + "/"
        + str(s.uuid)
        + "/"
        + s.unit
        + "/"
        + s.annotations()[0].get("type", " ")
    )


def process_collection(coll: str, ns_delta: int, choice: str = "latest"):
    """Processing pipeline for the collection"""
    df = get_streamset_and_values(coll, ns_delta, choice)
    res_dict = dict()
    res_dict["num_samples"] = df.apply(lambda x: x.shape[0])
    res_dict["num_unique_samples"] = df.apply(lambda x: x.unique().shape[0])
    res_dict["percent_unique_samples"] = df.apply(
        lambda x: x.unique().shape[0] / x.shape[0] * 100
    )
    #     res_dict['unique_values'] = df.apply(lambda x: x.unique().tolist())
    #     res_dict['unique_diff_values'] = df.apply(lambda x: x.diff().dropna().unique().tolist())
    res_dict["num_unique_diff_values"] = df.apply(
        lambda x: x.diff().dropna().unique().shape[0]
    )
    res_dict["percent_unique_diff_values"] = df.apply(
        lambda x: x.diff().dropna().unique().shape[0] / x.diff().dropna().shape[0] * 100
    )
    res_dict["num_decimals_mean"] = df.apply(
        lambda x: x.astype("string")
        .apply(decimal.Decimal)
        .apply(lambda y: abs(y.normalize().as_tuple().exponent))
    ).apply(np.mean)
    res_dict["num_decimals_min"] = df.apply(
        lambda x: x.astype("string")
        .apply(decimal.Decimal)
        .apply(lambda y: abs(y.normalize().as_tuple().exponent))
    ).apply(np.min)
    res_dict["num_decimals_max"] = df.apply(
        lambda x: x.astype("string")
        .apply(decimal.Decimal)
        .apply(lambda y: abs(y.normalize().as_tuple().exponent))
    ).apply(np.max)
    res_dict["num_decimals_std"] = df.apply(
        lambda x: x.astype("string")
        .apply(decimal.Decimal)
        .apply(lambda y: abs(y.normalize().as_tuple().exponent))
    ).apply(np.std)
    return res_dict
```
